## zfill()


*   Can add a numeric value of 0 to the left of a number or string
*   The number of characters must be less than or equal to the zfill value



```python
a = 7
```

```python
str(a).zfill(5)
```

```python
b = "200"
```

```python
b.zfill(5)
```

Implementation zfill() on strings

```python
a = "dia"
```

```python
a.zfill(6)
```

```python
b = "you"
```

```python
b.zfill(6)
```

```python
b.zfill(-1)
```

## Allignment
Align text right/left/center with rjust(), ljust(), and center()

### rjust()
- The rjust() method is used to make the text right-align
- Add spaces to strings to make them match
- the parameter is an integer which is the overall test length (not the number added)

```python
"Data".rjust(10)
```

```python
a = "Data".rjust(80)
```

```python
len(a)
```

### ljust()
- The rjust() method is used to make the text left-aligned
- Add spaces to strings to make them match
- the parameter is an integer which is the overall test length (not the number added)

```python
"Data".ljust(10)
```

### center()
- Make text centered
- The second parameter is used to replace the space

```python
"Data Engineer".center(20)
```

```python
"Data Engineer".center(20,"-")
```

## Literals String
A String Literal is a sequence of characters surrounded by quotation marks. We can use one, two, or three quotes for strings. Also, character literals are single characters surrounded by one or two quotes.

```python
"Jum''at"
```

Escape character is a backslash that allows inserting the characters ' and " in the string part

```python
"Jum\'at"
```

```python
print(f"Hello\nWorld!")
```

##Raw String
Python also provides a way to print strings according to any given input. Insert the letter r before the string opener

```python
print("Data\tScience")
```

```python
print(r"Data\tScience")
```
