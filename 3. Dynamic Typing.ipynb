#Dynamic Typing

### type()

Type() is one of the Python's built-in functions that can find the class type of the variable given as input.

```python
a = 20
type(a)
```

### len()
Returns the length of an object. For example, it can return the number of character in a string include white space, or return the number of element in list/set data type. It can be used with many different data types. However, not all data types are valid arguments for len().

```python
a = "Data Science"
len(a)
```

```python
b = 8
len(b)
```
