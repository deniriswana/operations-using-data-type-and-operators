### Range
- The range() function gives a series of numbers with a certain pattern
- For looping, for example, in accessing list elements, you can use the rane90 function in Python
- The range() function can have 1-3 parameters

```python
for i in range(3):
  print(i)
```

- range dwith 2 parameters n,p; make a series of numbers starting from n, until before p (number p does not follow)
- Often referred to as inclusive n (series starting from n) and exclusive (p is not included)

```python
for i in range(2,9):
  print(i)
```

```python
for i in range(2,9,2):
  print(i)
```
